```python
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)
```

```python
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
```

```python
tarih = df[(df["Date"] >= "2023-01-01") & (df["Date"] <= "2023-12-31")]
tarih
```

```python
m4_osmaniye = df[(df["Magnitude"] >= 4) &
                 (df["Location"].str.contains("Osmaniye", case=False, na=False))]
m4_osmaniye
```

```python
import matplotlib.pyplot as plt
```

```python
top10 = df["Location"].value_counts().head(10)
```

```python
plt.figure(figsize=(10,5))

plt.bar(top10.index, top10.values)

plt.title("En Çok Deprem Olan 10 Yer")
plt.xlabel("Yer")
plt.ylabel("Deprem Sayısı")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.bar(top10.index, top10.values)

plt.title("En Çok Deprem Olan 10 Yer", fontsize=14)
plt.xlabel("Yer")
plt.ylabel("Deprem Sayısı")

plt.xticks(rotation=45)
plt.grid(axis="y")

plt.tight_layout()

plt.savefig("en_cok_deprem_10_yer.png")

plt.show()
```

```python
yillar = df.groupby("Year").size()

plt.figure(figsize=(8,5))

plt.plot(yillar.index, yillar.values, marker="o")

plt.title("Yıllara Göre Deprem Sayısı")
plt.xlabel("Yıl")
plt.ylabel("Deprem Sayısı")

plt.grid(True)

plt.savefig("yillara_gore_deprem_sayisi.png")

plt.show()
```

```python
aylar = df["Month"].value_counts().sort_index()

plt.figure(figsize=(8,5))

plt.plot(aylar.index, aylar.values, marker="o")

plt.title("Aylara Göre Deprem Sayısı")
plt.xlabel("Ay")
plt.ylabel("Deprem Sayısı")

plt.xticks(range(1,13))

plt.grid(True)

plt.savefig("aylara_gore_deprem_sayisi.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["Magnitude"], bins=20)

plt.title("Deprem Büyüklüğü Dağılımı")
plt.xlabel("Büyüklük")
plt.ylabel("Deprem Sayısı")

plt.grid(True)

plt.savefig("buyukluk_histogram.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["Depth"], bins=20)

plt.title("Deprem Derinliği Dağılımı")
plt.xlabel("Derinlik (km)")
plt.ylabel("Deprem Sayısı")

plt.grid(True)

plt.savefig("derinlik_histogram.png")

plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(df["Longitude"], df["Latitude"], s=5)

plt.title("Depremlerin Enlem-Boylam Dağılımı")
plt.xlabel("Boylam")
plt.ylabel("Enlem")

plt.grid(True)

plt.savefig("enlem_boylam_scatter.png")

plt.show()
```
